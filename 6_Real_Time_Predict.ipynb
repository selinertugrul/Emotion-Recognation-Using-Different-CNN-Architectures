```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

# Modelinizi yükleyin
model = load_model("C:/Users/ugur-/Desktop/Modeller/Modeller/3_CNN_3/model3.h5")

# Duygu etiketleri
emotion_labels = ['Angry', 'Disgusted', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprised']

# Gerçek zamanlı yüz algılama ve duygu gösterimi için fonksiyon
def real_time_emotion_detection():
    face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0)  # Webcam aktif

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

        for (x, y, w, h) in faces:
            roi_gray = gray[y:y+h, x:x+w]
            roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)

            roi = roi_gray.astype('float') / 255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi, axis=0)

            # Duygu tahmini yapın+
            predictions = model.predict(roi)[0]
            max_index = np.argmax(predictions)
            emotion = emotion_labels[max_index]
            probability = predictions[max_index]

            # Tahmini görüntüye yazdırın
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
            cv2.putText(frame, emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 127, 255), 2)
            cv2.putText(frame, f"{probability:.2f}", (x + w - 50, y + h - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 127, 255), 2)

        # Görüntüyü gösterin
        cv2.imshow('Real-time Emotion Detector', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):  # 'q' tuşuyla çıkış
            break

    cap.release()
    cv2.destroyAllWindows()

# Gerçek zamanlı duygu tespiti başlat
real_time_emotion_detection()
```
