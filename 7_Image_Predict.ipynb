```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

# Modelinizi yükleyin
model = load_model("C:/Users/ugur-/Desktop/MODELLER SON/ResNET.h5")

# Duygu etiketleri
emotion_labels = ['Angry', 'Disgusted', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprised']

# Yüz algılama ve duygu gösterimi için fonksiyon
def detect_and_display(frame, color_mode='grayscale'):
    face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    # Görüntüyü belirtilen renk moduna göre ayarlayın
    if color_mode == 'grayscale':
        processed_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    else:
        processed_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Yüzleri algılayın
    faces = face_classifier.detectMultiScale(processed_img, scaleFactor=1.3, minNeighbors=5)
    
    for (x, y, w, h) in faces:
        # Yüz için ROI (Region of Interest) oluşturun
        roi = processed_img[y:y+h, x:x+w]
        roi = cv2.resize(roi, (48, 48), interpolation=cv2.INTER_AREA)
        
        if color_mode == 'grayscale':
            roi = roi.astype('float') / 255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi, axis=0)
        else:
            roi = roi.astype('float') / 255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi, axis=0)

        # Duygu tahmini yapın
        predictions = model.predict(roi)[0]
        max_index = np.argmax(predictions)
        emotion = emotion_labels[max_index]
        probability = predictions[max_index]
        
        # Tahmini görüntüye yazdırın
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 127, 255), 2)
        #cv2.putText(frame, f"{probability:.2f}", (x + w - 50, y + h - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 127, 255), 2)
    
    return frame

# Video işleme fonksiyonu
def process_video(video_path, output_path, color_mode='grayscale'):
    cap = cv2.VideoCapture(video_path)
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Her kareyi işleyin
        frame = detect_and_display(frame, color_mode)
        
        # İşlenen kareyi yazdırın
        out.write(frame)
        
        # (İsteğe bağlı) Her kareyi gösterin
        cv2.imshow('Emotion Detector', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Video yolu ve renk modu
video_path = "C:/Users/ugur-/Desktop/emotionvideos1.mp4"
output_path = "C:/Users/ugur-/Desktop/output_video12.mp4"
color_mode = 'grayscale'  # 'rgb' veya 'grayscale' olarak değiştirilebilir
process_video(video_path, output_path, color_mode)
```
